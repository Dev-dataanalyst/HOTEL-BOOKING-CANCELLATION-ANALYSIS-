## Data Analysis and Data Visualization

```python
# Check what percent of cacellation happened in past years

cancelled_perc = df['is_canceled'].value_counts(normalize = True)*100  #normalize will return values in percentage
cancelled_perc
```

#### Above result shows that from overall reservation only 62.8% is not cancelled reservation but cancellation percentage is 37% which is very high.

### Visualization of Cancelation

```python
sns.set_style('white')
plt.figure(figsize = (5,4))
plt.title('Reservation Status')
plt.bar(['Not Cancelled','Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'black', color = 'blue')
plt.ylabel('Number of reservation')
sns.despine()
plt.show()
```

#### Above graph shows that number of reservation is canceled very high and just 50% of total reservation.

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette='Blues',width = 0.7)
plt.legend(['not canceled','canceled'])
plt.title('Reservation status in different hotels')
plt.xlabel('hotel')
plt.ylabel('number of reservation')
```

#### Above graph shows that maximum cancelation in City hotel instead of Resort hotel

```python
# Let clear more above graph in percentage.

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

#### Let's check whether price also factor for cancelation.

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel.head()
```

```python
plt.figure(figsize=(20,8))
plt.plot(resort_hotel.index,resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.title("Average Daily Rate in City and Resort Hotel", fontsize=20, fontweight = 'bold')
plt.show()
```

#### Above graph shows that price of city hotel generally low in comparsion to resort hotel.

```python
# Check in which month reservation and cancelation is high

# extract month from reservation date column

df['month'] = df['reservation_status_date'].dt.month


plt.figure(figsize=(12,4))
ax1 = sns.countplot(x='month', hue = 'is_canceled', data = df, palette='bright')

plt.title('Reservation status per month', fontsize = 20, fontweight = 'bold')
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])

plt.show()
```

#### Above graph shows that in month of January most of the reservation canceled and low rate of cancelation in august month.

```python
# Check price rate also factor in particular month or not.

plt.figure(figsize=(15,5))
plt.title('ADR per month', fontsize = 20)
sns.barplot(x='month', y='adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
```

#### Above graph shows that maximum price in month of January month and lowest price in august month. This somehow proof that maximum cancelation due to price diferences.

```python
# now check country wise cancellation 

cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
```

```python
plt.figure(figsize=(8,5))
plt.pie(top_10_country,autopct='%.2f', labels=top_10_country.index)
plt.title('Top 10 countries with reservation cancelled')
plt.show()
```

#### Above pie chart clearly show that maximum cancellation of reservation in protugal country.

```python
# Now check wheter not canceled or canceled high in comparison to ADR (Average Daily Rate)

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()
cancelled_df_adr.sort_values('reservation_status_date')

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()
not_cancelled_data.sort_values('reservation_status_date')

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize = 20)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate (2016 - 2017) ', fontsize = 20)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
plt.show()
```
